```python
def grouper( x , idx ,col):
    if 80 <= x[col].loc[idx]<=90:
        return '80-90'
    elif 50 <= x[col].loc[idx]<=60:
        return '50-60'
    else:
        return 'others'
```

```python
g = x.groupby(lambda idx: grouper(x,idx,'temperature'))
g
```

```python
for key, d in g:
    print('key : ',key)
    print('Data : ',d)
```
