```python
df.head()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(12,7))
plt.plot(df["Quarter"],df["Fridge"],color="green")
plt.plot(df["Quarter"],df["Dishwasher"],color="red")
plt.plot(df["Quarter"],df["Washing Machine"],color="blue")
plt.title("Product Sales")
plt.xlabel("Financial Quarter")
plt.ylabel("Revenue(Min $ ) ")
plt.legend()
plt.show()
```

```python
total_sales = df[["Fridge","Dishwasher","Washing Machine"]].sum()
print(total_sales)
```

```python
plt.pie(total_sales,labels=total_sales.index,autopct="%1.1f%%",explode=(0.1,0.1,0.1 ),shadow=True)
plt.show()
```

```python
df.plot(kind = "bar",x="Quarter")
plt.xticks(rotation=45)
plt.show()
```
