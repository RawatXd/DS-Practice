```python
num = [1,2,3,4,5]
s = []
for n in num:
    s.append(n*n)

print(s)
```

```python
s = [n*n for n in num]
s
```

```python
import time

long_list =  range(10000000)
sqr_list = []

start = time.time()
for n in long_list:
    sqr_list.append(n*n)

end = time.time()
print("Time Taken by for LOOP : ",(end-start))
```

```python
start = time.time()
sqr_list = [n*n for n in long_list]
end = time.time()
print("Time Taken by for LOOP : ",(end-start))
```
