```python
from fontTools.merge import timer


def get_square(numbers):
    result = []
    for n in numbers:
        result.append(n*n)
    return result
```

```python
def get_cube(numbers):
    result = []
    for n in numbers:
        result.append(n*n*n)
    return result
```
